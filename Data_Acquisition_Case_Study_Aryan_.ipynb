```python
import pandas as pd
```

```python
#population (csv)
population = pd.read_csv('/content/population.csv')
population
```

```python
#GDP (excel)
GDP = pd.read_excel('/content/gdp.xlsx')
GDP
```

```python
#internet users (json)
internet_users = pd.read_json('/content/internet_users.json')
internet_users
```

```python
#literacy rate (xml)
literacy_rate = pd.read_xml('/content/literacy_rate.xml')
literacy_rate
```

```python
GDP.rename(columns={'Country':'country'},inplace=True)
literacy_rate.rename(columns={'name':'country'},inplace=True)
```

```python
# Merge the datasets based on the country name
merged_df = pd.merge(population,GDP,on='country',how='inner')
merged_df = pd.merge(merged_df,internet_users,on='country',how='inner')
merged_df = pd.merge(merged_df,literacy_rate,on='country',how='inner')
merged_df
```

```python
# Create new columns such as internet penetration rate (internet users / population × 100)
merged_df['internet_penetration_rate']=merged_df['internet_users']/merged_df['population']*100
merged_df
```

```python
# Handle unmatched or missing records ✓
# Validate the data for consistency and completeness ✓
```

```python
# Find countries with highest internet penetration rates
merged_df.sort_values(by='internet_penetration_rate',ascending=False).head(3)
```

```python
# Calculate average literacy rate across countries
merged_df['literacy_rate'].mean()
```

```python
# Investigate relationships between GDP, population, literacy, and internet usage
merged_df[['internet_users','GDP','population','literacy_rate']].corr()
```

```python
# Create an SQLite database
# Load the final merged dataset into a database table
import sqlite3
conn = sqlite3.connect('global_data.db')
merged_df.to_sql('global_data',conn,index=False,if_exists='replace')
```

```python
pd.read_sql_query('select * from global_data',conn)
```

```python
# e. Which countries have a GDP per capita above $10,000?

cursor = conn.cursor()
#Add the new column
cursor.execute("ALTER TABLE global_data ADD COLUMN GDP_per_capita REAL")

```

```python
cursor.execute("""
  UPDATE global_data
  SET GDP_per_capita = GDP * 1.0/population
  """)
conn.commit()
```

```python
# e. Which countries have a GDP per capita above $10,000?
pd.read_sql_query('select country,GDP_per_capita from global_data where GDP_per_capita > 10000',conn)
```

```python
# f. What is the total population covered in the dataset?
pd.read_sql_query('select sum(population) as total_population from global_data',conn)
```

```python
# g. Which countries have the lowest literacy rates, and how does that impact internet access?
pd.read_sql_query('select country, literacy_rate, internet_penetration_rate from global_data order by literacy_rate limit(5)',conn)
```

```python
# h. What are the top 5 wealthiest countries by total GDP, and how does that compare with population size?
pd.read_sql_query('select country, GDP from global_data order by GDP desc limit 5 ',conn)
```

```python
# i. Find countries where internet users exceed 70% of the population.
pd.read_sql_query('select country from global_data where internet_users > 0.7 * population', conn)
```

```python
# j. What is the average GDP per capita for countries with internet penetration above 50%?
pd.read_sql_query('''select avg(gdp_per_capita)
                     from (select gdp/population as gdp_per_capita
                           from global_data where internet_penetration_rate > 50)''', conn)
```
