```python
import pandas as pd
import numpy as np
# import matplotlib
# import sklearn
# import scipy
import seaborn as sns
```

```python
netflix_df = pd.read_excel("/content/partpdf_1763620928546_netflix_titles.xlsx",sheet_name=None)
```

```python
netflix_titles = netflix_df['netflix_titles']
netflix_titles_directors = netflix_df['netflix_titles_directors']
netflix_titles_countries= netflix_df['netflix_titles_countries']
netflix_titles_cast =netflix_df['netflix_titles_cast']
netflix_titles_category = netflix_df['netflix_titles_category']
```

```python
merged_df = netflix_titles.copy()
dataframes_to_merge = [netflix_titles_directors, netflix_titles_countries, netflix_titles_cast, netflix_titles_category]

for current_df in dataframes_to_merge:
    merged_df = merged_df.merge(current_df, on="show_id", how="left")

merged_df.info()
```

```python
# Merge Strategy

# To combine the main `netflix_titles` file with the additional sheets, I used a **left join** on the `show_id` column for each merge operation.
# The left join is appropriate because the goal is to ensure all entries in the main titles dataset appear in the final master DataFrame,
# even if some do not have matching data in the supplementary sheets (e.g.,netflix_titles,netflix_titles_directors, netflix_titles_cast, netflix_titles_category)
# If I used an inner join, any title missing related data in any sheet would be excluded, which is not desired for comprehensive analysis.
# Left join preserves all titles and adds supplementary info when available, filling gaps with NaN.
```

```python
# Fill missing director and country data
merged_df['director'] = merged_df['director'].fillna('Unknown')
merged_df['country'] = merged_df['country'].fillna('Unknown')

# Handle missing date_added and convert to datetime
merged_df['date_added'] = pd.to_datetime(merged_df['date_added'], errors='coerce')
# Create year_added column
merged_df['year_added'] = merged_df['date_added'].dt.year

# If you want to see the info output for screenshot:
merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# Select relevant columns (ensure date_added is datetime and year_added is numeric)
cols = ['duration_minutes', 'duration_seasons', 'date_added', 'year_added']

# Convert duration columns to numeric, coercing errors to NaN
merged_df['duration_minutes'] = pd.to_numeric(merged_df['duration_minutes'], errors='coerce')
merged_df['duration_seasons'] = pd.to_numeric(merged_df['duration_seasons'], errors='coerce')

# Convert date_added to numeric (e.g., timestamp) for skewness calculation
merged_df = merged_df.copy()
merged_df['date_added'] = pd.to_datetime(merged_df['date_added'], errors='coerce')
merged_df['date_added_numeric'] = merged_df['date_added'].map(lambda x: x.timestamp() if pd.notnull(x) else None)

# Prepare a dictionary of column names and data to analyze skewness
skew_data = {
    'duration_minutes': merged_df['duration_minutes'].dropna(),
    'duration_seasons': merged_df['duration_seasons'].dropna(),
    'date_added': merged_df['date_added_numeric'].dropna(),
    'year_added': merged_df['year_added'].dropna()
}

# Calculate skewness
skewness = {col: skew(data) for col, data in skew_data.items()}

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(skewness.keys(), skewness.values(), color='skyblue')
plt.axhline(0, color='black', linewidth=0.5)
plt.title("Skewness of Selected Columns")
plt.ylabel("Skewness")
plt.show()
```

```python
# Calculate the median of the column, ignoring nulls
median_seasons = merged_df['duration_seasons'].median()

# Fill nulls with median
merged_df['duration_seasons'] = merged_df['duration_seasons'].fillna(median_seasons)

# For 'date_added', first convert to datetime, then to ordinal/int (if needed)
# If already numeric, skip conversion step
merged_df['date_added'] = pd.to_datetime(merged_df['date_added'], errors='coerce')
# Fill missing 'duration_minutes' with mean
merged_df['duration_minutes'] = merged_df['duration_minutes'].fillna(merged_df['duration_minutes'].mean())

# Fill missing 'year_added' with mean (rounded to int if desired)
merged_df['year_added'] = merged_df['year_added'].fillna(merged_df['year_added'].mean())

# For 'date_added', fillna with mean date
mean_date = pd.to_datetime(np.nanmean(merged_df['date_added'].dropna().astype(np.int64)), unit='ns')
merged_df['date_added'] = merged_df['date_added'].fillna(mean_date)

# drop unnecessary col
merged_df = merged_df.drop('date_added_numeric', axis=1)
```

```python
# **Strategy for Handling Missing Country Data:**

# **Strategy for Handling Missing Country Data:**rather than dropping rows or using more
# complex imputation methods. This approach ensures that we retain all titles for analysis,
# especially as country information may be absent due to incomplete metadata rather than true
# absence. Dropping rows would risk losing valuable data and skewing analysis, while advanced
# imputation may introduce bias. Using `'Unknown'` provides a clear marker for missing data,
# keeping the dataset intact
# i also filled skewd vals with median
# and non skewd vals with mean
# and i thought its not apropriate in my case to use mode to fill null categorical columns,as it might adultrate the dataframe
```

```python
merged_cast = merged_df.dropna(subset=['cast']).copy()

# Split 'cast' strings into lists, then explode
merged_cast['cast'] = merged_cast['cast'].str.split(', ')
exploded_cast = merged_cast.explode('cast').reset_index(drop=True)

print(exploded_cast[['show_id', 'cast']].head())
```

```python
# **Business Advantage of â€œExplodingâ€ the Cast Column:**

# By transforming the â€˜castâ€™ column so that each actor gets their own row, the data becomes much more flexible and powerful for analysis. Now, we can:

# - Directly count the appearances of each actor across all titles to find the most frequent actors on the platform.
# - Analyze trends, such as how often certain actors appear in Netflixâ€™s content by year, genre, or country.
# - Build actor-specific dashboards, recommendations, or partnership strategies.
# - Identify collaborative patterns, such as which actors often co-star together.

# Prior to this transformation, such insights were difficult or even impossible to compute efficiently because all actors were stored together in a single string for each show.
```

```python
import matplotlib.pyplot as plt

# Count actor occurrences
top_actors = exploded_cast['cast'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
top_actors.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Frequent Actors on Netflix')
plt.xlabel('Actor')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# The bar chart reveals that the most frequent actors on Netflix are heavily concentrated
# among a few performers. For example, Actor XYZ appears most often, indicating their
# significant contribution to the streaming serviceâ€™s catalog. This suggests Netflix may
# prioritize working with popular or prolific actors to attract and retain viewers, leveraging
# star power for successful releases.
```
