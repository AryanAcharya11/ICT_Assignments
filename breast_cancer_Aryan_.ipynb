```python
# Import necessary libraries for data manipulation, visualization, and machine learning.

import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
# Load the Breast Cancer dataset and create a pandas DataFrame.
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

cancer_df = pd.DataFrame(data.data, columns=data.feature_names)
cancer_df['target'] = data.target

cancer_df.head()
```

```python
cancer_df.to_csv('bcdata.csv')
```

```python
cancer_df.describe()
```

```python
cancer_df.info()
```

```python
y = cancer_df.target
x = cancer_df.drop('target', axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('scaler', RobustScaler(), x.columns)
    ]
)
```

```python
lr_model = LogisticRegression(random_state=42)
dtc_model = DecisionTreeClassifier(random_state=42)
rfc_model = RandomForestClassifier(random_state=42)
xgbc_model = XGBClassifier(random_state=42)
```

```python
models = [lr_model, dtc_model, rfc_model, xgbc_model]
```

```python
for model in models:
  pipeline = Pipeline(steps=[
  ('preprocessor', preprocessor),
  ('model', model)
  ])

  pipeline.fit(x_train, y_train)
  y_pred = pipeline.predict(x_test)
  print(f'Model: {type(model).__name__}')
  print(f'Accuracy score: {accuracy_score(y_test, y_pred):.4f}')
```

```python
pca_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('pca', PCA())
])

x_processed_pca = pca_pipeline.fit_transform(x_train)
pca = pca_pipeline.named_steps['pca']
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulated explained variance')
plt.title('Components vs. Explained variance')
plt.grid(True)
plt.show()
```

```python
x_preprocessed = preprocessor.fit_transform(x_train)
pca = PCA(n_components=0.95, random_state=42)
x_pca = pca.fit_transform(x_preprocessed)
print('chosen components: ', pca.n_components_)
```

```python
for model in models:
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('pca', PCA(n_components=10, random_state=42)),
        ('model', model)
    ])
    pipeline.fit(x_train, y_train)
    y_pred = pipeline.predict(x_test)
    print(f"Model: {type(model).__name__}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
```
