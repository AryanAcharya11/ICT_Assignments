```python
import pandas as pd
import numpy as np
```

```python
# load data
earthquke_df = pd.read_csv("https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv")
```

```python
# identify number of records & unique locations

print(f'Records: {len(earthquke_df)}')
print(f"Unique locations: {earthquke_df['place'].nunique()}")
```

```python
# 3.Print top 5 rows and column names

print('First five rows: ')
earthquke_df.head()
```

```python
print(f"Column names: {earthquke_df.columns.tolist()}")
```

```python
# 4. Convert time to datetime

earthquke_df['time'] = pd.to_datetime(earthquke_df['time'])
```

```python
# 5. Drop records with missing latitude, longitude, mag

earthquke_df = earthquke_df.dropna(subset = ['latitude', 'longitude', 'mag'])
```

```python
# 6.  Filter only earthquakes with magnitude >= 4.0

earthquke_df[earthquke_df['mag'] > 4]
```

```python
# 7. Add a column day_of_week from time

earthquke_df['day_of_week'] = earthquke_df['time'].dt.day_name()
```

```python
# 8. Create a column severity_level based on magnitude:
# ○ <4.0: "Low"
# ○ 4.0 - 6.0: "Moderate"
# ○ 6.0+: "High"

conditions = [earthquke_df['mag'] < 4,
              earthquke_df['mag'].between(4, 6),
              earthquke_df['mag'] > 6]
choices = ['Low', 'Moderate', 'High']

earthquke_df['severity_level'] = np.select(conditions, choices, default='Unknown')
```

```python
earthquke_df
```

```python
# 9. Count number of earthquakes per place

earthquke_df['place'].value_counts()
```

```python
# 10. Compute average magnitude and max depth per day

earthquake_summary = earthquke_df.groupby(earthquke_df['time'].dt.date).agg({'depth':'max', 'mag':'mean'}).reset_index()
earthquake_summary.columns = ['date', 'max_depth', 'average_magnitude']
```

```python
earthquake_summary
```

```python
# 11. Save:
# ○ Cleaned dataset to cleaned_earthquakes.csv
# ○ Summary data to earthquake_summary.csv

earthquke_df.to_csv('cleaned_earthquakes.csv')
earthquake_summary.to_csv('earthquake_summary.csv')
```
