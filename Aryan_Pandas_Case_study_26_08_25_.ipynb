```python
# 4) Check and remove duplicates
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
# 5) Rename columns ("Title" > "Movie_Title")
```

```python
df.rename(columns={'Title':'Movie_Title'}, inplace=True)
df.head()
```

```python
# 6) Get unique genres
```

```python
unique_genres = df['Genre'].unique()
print("\nUnique Genres:", unique_genres)
```

```python
# 7) Extract number of genres per movie
```

```python
df['Num_Genres'] = df['Genre'].apply(lambda x: len(x.split(',')))
df
```

```python
# 8) Handle missing values in 'Revenue (Millions)' and 'Metascore
```

```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].median())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].median())
df
```

```python
# 9) Top 5 movies by Rating
```

```python
print("\nTop 5 Movies by Rating:\n", df.nlargest(5, 'Rating')[['Movie_Title','Rating']])
```

```python
# 10) Movies with highest/lowest revenue
```

```python
highest_revenue = df.loc[df['Revenue (Millions)'].idxmax()]
lowest_revenue = df.loc[df['Revenue (Millions)'].idxmin()]
print("\nHighest Revenue Movie:\n", highest_revenue[['Movie_Title','Revenue (Millions)']])
print("\nLowest Revenue Movie:\n", lowest_revenue[['Movie_Title','Revenue (Millions)']])
```

```python
# 11) Avg. rating by Director
```

```python
print("\nAverage Rating by Director:\n", df.groupby('Director')['Rating'].mean().head())
```

```python
# 12) Movies per year
```

```python
print("\nMovies per Year:\n", df['Year'].value_counts().sort_index())
```

```python
# 13) Movies with rating > 8 & revenue > 100M
```

```python
print("\nHigh Rated & High Revenue Movies:\n",
      df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)][['Movie_Title','Rating','Revenue (Millions)']])
```

```python
# 14) Group by Director: avg Rating, Revenue
print("\nDirector-wise Avg Rating & Revenue:\n",
      df.groupby('Director')[['Rating','Revenue (Millions)']].mean().head())
```

```python
# 15) Group by Year: count, avg Rating
print("\nYearly Count & Avg Rating:\n",
      df.groupby('Year').agg({'Movie_Title':'count','Rating':'mean'}).rename(columns={'Movie_Title':'Movie_Count'}))
```

```python
# Create new column: Revenue divided by Duration
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']

# Display first few rows to confirm
print(df[['Movie_Title', 'Revenue (Millions)', 'Runtime (Minutes)', 'Profitability']].head())
```

```python
# 17) Categorize movies by rating
def categorize_rating(r):
    if r < 6:
        return "Poor"
    elif 6 <= r < 7.5:
        return "Average"
    else:
        return "Excellent"

df['Rating_Category'] = df['Rating'].apply(categorize_rating)
```

```python
# 18) Custom function to score movies
def high_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['High_Score'] = df.apply(high_score, axis=1)

print("\nDataset with new columns:\n", df.head())
```
